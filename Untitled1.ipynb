```python
import pybaseball
import os 
import pandas as pd 
import numpy as np 
import itertools
from tqdm.notebook import tqdm
import time
import pickle

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, log_loss, accuracy_score
from scipy.stats import pearsonr, gaussian_kde, spearmanr
from scipy.signal import argrelextrema, find_peaks
import lightgbm as lgb

import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df_with_re = pd.read_csv("data/20182019_data_w_pred.csv")
```

```python
asv_shift = df_with_re['actual_stuff_value'].mean() - df_with_re['pitch_level_runs'].mean()
df_with_re['vacuum_val'] = df_with_re['vacuum_val'].sub(asv_shift/2)
df_with_re['swstr_val'] = df_with_re['swstr_val'].sub(asv_shift/2)
df_with_re['actual_stuff_value'] = (df_with_re['vacuum_val']).add(df_with_re['swstr_val'])

groupby_cols = ['game_year','player_name', 'pitcher']
smol_feats1 = ['game_year','player_name', 'pitcher', 'pitch_level_runs', 'vacuum_val', 
               'swstr_val', 'actual_stuff_value']
               
df_weight = df_with_re.loc[:, smol_feats1].groupby(groupby_cols, as_index=False).sum()
df_weight['roe'] = df_weight['pitch_level_runs'].sub(df_weight['actual_stuff_value'])
loc_r2 = spearmanr(df_weight['vacuum_val'], df_weight['roe'])[0]**2
move_r2 = spearmanr(df_weight['swstr_val'], df_weight['roe'])[0]**2
movement_weight = move_r2/loc_r2

df_with_re['pitch_value'] = (df_with_re['vacuum_val']).add(df_with_re['swstr_val'].mul(movement_weight))
df_with_re['roe'] = df_with_re['pitch_level_runs'].sub(df_with_re['pitch_value'])
```

```python
target_year = 2019
groupby_cols = ['player_name', 'pitcher']
smol_feats1 = ['player_name', 'pitcher', 'pitch_level_runs', 'vacuum_val', 
               'swstr_val', 'actual_stuff_value', 'pitch_value']

smol_feats2 = ['player_name', 'pitcher', 'estimated_woba_using_speedangle', 
              'woba_value', 'woba_denom', 'babip_value', 'iso_value']              
_df_corr = df_with_re.loc[df_with_re['game_year']==target_year, smol_feats1].groupby(groupby_cols, as_index=False).sum()
df_corr = _df_corr.merge(df_with_re.loc[df_with_re['game_year']==target_year, smol_feats2].groupby(groupby_cols, as_index=False).mean())
df_corr['roe'] = df_corr['pitch_level_runs'].sub(df_corr['actual_stuff_value'])

player_lkup = pybaseball.playerid_reverse_lookup(df_with_re['pitcher'].unique())
player_lkup = player_lkup.loc[player_lkup['key_fangraphs']!=-1, ['key_mlbam', 'key_fangraphs']].drop_duplicates()
player_lkup = player_lkup.rename(columns={'key_fangraphs':'IDfg', 'key_mlbam':'pitcher'})

fg = pybaseball.fg_pitching_data(target_year)

df_corr = player_lkup.merge(fg).merge(df_corr)
```

```python
rank_by = 'roe'
threshold_n = 40
threshold_mix = .05
threshold = (df_with_re
 .loc[:, ['game_year','pitcher','player_name', 'pitch_name', 'pitch_type', 're288_delta']]
 .groupby(['game_year','pitcher','player_name', 'pitch_name', 'pitch_type'], 
          as_index=False)
 .count().rename(columns={'re288_delta':'times_thrown'}))

threshold = (threshold
             .merge(threshold
                    .loc[:, ['game_year','pitcher','times_thrown']]
                    .groupby(['game_year','pitcher'], 
                             as_index=False)
                    .sum()
                    .rename(columns={'times_thrown':'total_pitches_thrown'})))

threshold['mix%'] = threshold['times_thrown'].div(threshold['total_pitches_thrown'])
threshold = (threshold
             .loc[(threshold['times_thrown']>=threshold_n) & 
                  (threshold['mix%']>=threshold_mix)])

pitch_ranks = (threshold
 .merge(df_with_re
        .loc[:, ['game_year','pitcher','player_name', 'pitch_name', 'pitch_type', 
                 're288_delta', 'pitch_level_runs', 'vacuum_val', 'swstr_val', 
                 'pitch_value', 'roe']]
        .groupby(['game_year','pitcher','player_name', 'pitch_name', 'pitch_type'], 
                 as_index=False)
        .mean())
 .sort_values(rank_by)).reset_index(drop=True)

model = LinearRegression()
model.fit(pitch_ranks.loc[:, ['vacuum_val']], pitch_ranks.loc[:, 'swstr_val'])
pitch_ranks['_split'] = (pd.Series(model.predict(pitch_ranks[['vacuum_val']]), 
                                   index=pitch_ranks.index) > pitch_ranks['swstr_val'])

model = LinearRegression()
model.fit(pitch_ranks.loc[pitch_ranks['_split']==1, ['vacuum_val']], pitch_ranks.loc[pitch_ranks['_split']==1, 'swstr_val'])
pitch_ranks['tier1'] = pd.Series(model.predict(pitch_ranks[['vacuum_val']]), index=pitch_ranks.index) > pitch_ranks['swstr_val']

model = LinearRegression()
model.fit(pitch_ranks.loc[pitch_ranks['_split']==0, ['vacuum_val']], pitch_ranks.loc[pitch_ranks['_split']==0, 'swstr_val'])
pitch_ranks['tier3'] = pd.Series(model.predict(pitch_ranks[['vacuum_val']]), index=pitch_ranks.index) < pitch_ranks['swstr_val']

pitch_ranks['tier'] = 0
pitch_ranks.loc[(pitch_ranks['tier1']==pitch_ranks['tier3']), 'tier'] = 2
pitch_ranks.loc[(pitch_ranks['tier1']==True), 'tier'] = 1
pitch_ranks.loc[(pitch_ranks['tier3']==True), 'tier'] = 3
pitch_ranks = pitch_ranks.drop(['tier1', 'tier3'], axis=1)

_pitch_ranks = pd.DataFrame()
for yr in range(2018,2020):
  ps = pybaseball.pitching_stats_bref(yr)
  _df = pitch_ranks.loc[pitch_ranks['game_year']==yr]
  _df = _df.merge(ps.loc[:, ['Name', 'G', 'GS']].rename(columns={'Name':'player_name'}))
  _df['gs%'] = _df['GS'].div(_df['G'])
  _pitch_ranks = _pitch_ranks.append(_df)
pitch_ranks = _pitch_ranks
pitch_ranks = pitch_ranks.sort_values(rank_by).reset_index(drop=True)
pitch_ranks['rank'] = range(1,pitch_ranks.shape[0]+1)

pitch_map = {}
cmap = sns.palettes.color_palette('tab10', n_colors=pitch_ranks['pitch_type'].unique().shape[0], as_cmap=False)
for x, y in zip(pitch_ranks['pitch_type'].unique(), cmap):
  pitch_map = {**pitch_map, **{x:y}}
```

```python
def pitcher_overall_value(player_name, df):
    #year_agg the next three lines should be a table that is precalculated from df_with_re
    year_agg_columns = ['game_year', 'player_name', 'pitcher', 're288_delta',
                        'pitch_level_runs','vacuum_val','swstr_val','pitch_value', 'roe']
    year_agg_groupby_columns = ['game_year', 'player_name','pitcher']
    year_agg = df.loc[:, year_agg_columns].groupby(year_agg_groupby_columns, as_index=False).sum()
    player_year_agg = year_agg.loc[year_agg['player_name']==player_name, 
                                   ['game_year', 're288_delta','pitch_level_runs','vacuum_val',
                                    'swstr_val','pitch_value', 'roe']].groupby(['game_year'], 
                                                                               as_index=False).sum()
    player_year_agg['hue'] = (player_year_agg['roe']<0).astype(int)

    cmap_dict = {0:(142/255, 0/255, 181/255), 1:(252/255, 186/255, 3/255)}
    cmap = []
    if player_year_agg['hue'].unique().shape[0] < 2:
        cmap = [cmap_dict[player_year_agg['hue'].unique()[0]]]
        hue_order = [player_year_agg['hue'].unique()[0]]
    else:
        for x in player_year_agg['hue'].unique():
            cmap.append(cmap_dict[x])
        hue_order = [0,1]

    fig = sns.scatterplot(x='pitch_level_runs', y='pitch_value', data=year_agg, alpha=.2, color=(.75,.75,.75), zorder=5)
    fig = sns.scatterplot(x='pitch_level_runs', y='pitch_value', data=player_year_agg, zorder=6, hue='hue', hue_order=hue_order, palette=cmap, legend=None)
    fig.set_aspect('equal')
    x_lim = fig.get_xlim()
    y_lim = fig.get_ylim()
    plt.plot([x_lim[0],x_lim[1]], [x_lim[0],x_lim[1]], 'k:', zorder=7)
    plt.title(f"{player_name} overall pitcher value")
    sns.despine()
    plt.show()
    df_export = player_year_agg.drop('hue', axis=1).sort_values('roe').reset_index(drop=True)
    return df_export
```
